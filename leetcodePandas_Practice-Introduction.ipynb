# General Practice - 1. Create a DataFrame from List

```python
student_data = [[1, 15], [2, 11], [3, 11], [4, 20]]

df = pd.DataFrame(student_data,)
# df.rename(columns={0:'student_id', 1:'age'}) # this is alternate way
df.columns = ["student_id", "age"]
df
```

## leetcode

# [1. Create a DataFrame from List](https://leetcode.com/problems/create-a-dataframe-from-list/description/?envType=study-plan-v2&envId=introduction-to-pandas&lang=pythondata)

```python
# 1. Create a DataFrame from List

def createDataframe(student_data) -> pd.DataFrame:
    column_names = ['student_id','age']
    df = pd.DataFrame(student_data, columns = column_names)
    return df

student_data = [[1, 15], [2, 11], [3, 11], [4, 20]]
createDataframe(student_data)
```

# General Practice - 6. Drop Duplicate Rows

```python
excel_file = pd.ExcelFile("general_practice.xlsx")
customers = pd.read_excel(excel_file, 'customers')
customers
```

```python
# Write a solution to remove these duplicate rows and keep only the first occurrence.
customers.drop_duplicates(subset = ['email'], keep='first', inplace = True)
customers
```

## leetcode

# [6. Drop Duplicate Rows](https://leetcode.com/problems/drop-duplicate-rows/?envType=study-plan-v2&envId=introduction-to-pandas&lang=pythondata)

```python
# 6. Drop Duplicate Rows
def dropDuplicateEmails(customers):
    df = pd.DataFrame(customers)
    df.drop_duplicates(subset=['email'], keep='first', inplace=True)
    return df

dropDuplicateEmails(customers)
```

# General Practice - 7. Drop missing data

```python
excel_file_name = "general_practice.xlsx"
sheet = "students"
excel_file = pd.ExcelFile(excel_file_name)
students = pd.read_excel(excel_file_name, sheet)
students
```

```python
# Write a solution to remove the rows with missing values.in column name
# if students['name'].isna:
students.dropna(subset=['name'],inplace=True)
students
```

## leetcode

# [7. Drop Missing Data](https://leetcode.com/problems/drop-missing-data/description/?envType=study-plan-v2&envId=introduction-to-pandas&lang=pythondata)

```python
def dropMissingData(students: pd.DataFrame) -> pd.DataFrame:
    df = pd.DataFrame(students)
    df.dropna(subset=['name'],inplace=True)
    return df

dropMissingData(students)
```

# General Practice 9. Rename Columns

```python
excel_file_name = "general_practice.xlsx"
sheet = 'new_students'
exce_file = pd.ExcelFile(excel_file_name)
students = pd.read_excel(excel_file, sheet)
students
```

Write a solution to rename the columns as follows:

- id to student_id
- first to first_name
- last to last_name
- age to age_in_years
The result format is in the following example.

```python
students.rename(columns = {"id":"student_id","first":"first_name", "last":"last_name", "age":"age_in_years"})
```

## leetcode

# [9. Rename Columns](https://leetcode.com/problems/rename-columns/?envType=study-plan-v2&envId=introduction-to-pandas&lang=pythondata)

```python
def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    df = pd.DataFrame(students)
    return students.rename(columns = {"id":"student_id","first":"first_name", "last":"last_name", "age":"age_in_years"})

renameColumns(students)
```

# General Practice 11. Fill Missing Data

```python
file_name = "general_practice.xlsx"
sheet = "products"

products = pd.read_excel(file_name, sheet)
products
```

Write a solution to fill in the missing value as 0 in the quantity column.

```python
products['quantity'] = products['quantity'].fillna(0)
products
```

## leetcode

# [11. Fill Missing Data](https://leetcode.com/problems/fill-missing-data/description/?envType=study-plan-v2&envId=introduction-to-pandas&lang=pythondata)

```python
def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    products['quantity'] = products['quantity'].fillna(0)
    return products

fillMissingValues(products)
```
